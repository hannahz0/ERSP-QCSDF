```python
import pandas as pd

column_names = ["query_id", "Q0", "doc_id", "rank", "score", "run_name"]

splade_df = pd.read_csv("splade-dev.trec", sep="\t", names=column_names)
bm25_df = pd.read_csv("bm25-t5-dev.trec", sep="\t", names=column_names)
```

```python
column_names_qrels = ["query_id", "zero", "doc_id", "relevance_score"]

qrels_df = pd.read_csv("qrels.dev.tsv", sep="\t", header=None, names=column_names_qrels)

qrels_dict = {}
for _, row in qrels_df.iterrows():
    query_id = row["query_id"]
    zero = row["zero"]
    doc_id = row["doc_id"]
    relevance_score = row["relevance_score"]

    if query_id not in qrels_dict:
        qrels_dict[query_id] = {}

    qrels_dict[query_id][doc_id] = relevance_score
    
qrels_df.head()
```

```python
column_names_queries = ["query_id", "query"]

queries_df = pd.read_csv("queries.dev.tsv", sep="\t", header=None, names=column_names_queries)
queries_df.head()
```

```python
queries = []

for query in queries_df["query"]:
    queries.append(query)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForMaskedLM
from splade.models.transformer_rep import Splade, SiameseBase

model_type = "naver/splade-cocondenser-ensembledistil"

model = SiameseBase(model_type_or_dir=model_type, output='cls', model_type_or_dir_q=model_type)

# print(model)
# print(model.transformer_rep_q)

model.eval()

tokenizer = AutoTokenizer.from_pretrained(model_type)
reverse_vocabulary = {v: k for k, v in tokenizer.vocab.items()}

# example document from MS MARCO passage collection (doc_id = 8003157)
ex_doc = "Glass and Thermal Stress. Thermal Stress is created when one area of a glass pane gets hotter than an adjacent area. If the stress is too great then the glass will crack. The stress level at which the glass will break is governed by several factors."
```

```python
# inputs = tokenizer(ex_doc, return_tensors="pt")

# with torch.no_grad():
#     outputs = model.transformer_rep_q(**inputs)  
    
# cls_token_representation = outputs.last_hidden_state[:, 0, :]

# print(cls_token_representation.shape)
```

```python
with torch.no_grad():
    doc_rep = model(d_kwargs=tokenizer(ex_doc, return_tensors="pt"))["d_rep"].squeeze()

col = torch.nonzero(doc_rep).squeeze().cpu().tolist()
print("number of actual dimensions: ", len(col))

# bag of words
weights = doc_rep[col].cpu().tolist()
d = {k: v for k, v in zip(col, weights)}
sorted_d = {k: v for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)}
bow_rep = []
for k, v in sorted_d.items():
    bow_rep.append((reverse_vocabulary[k], round(v, 2)))
print("SPLADE BOW rep:\n", bow_rep)
```

```python
cls_tokens = []

for query in queries:
    if model.transformer_rep_q is not None:
        model.transformer_rep_q.output = "cls"
        tokens = tokenizer(query, return_tensors="pt")
        cls = model.transformer_rep_q.forward(**tokens)
        cls_tokens.append(cls)
        model.transformer_rep_q.output = "MLM"


    else:
        print("Error: transformer_rep_q is not initialized")
```
